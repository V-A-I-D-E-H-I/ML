```python
import pandas as pd
import numpy as np
import seaborn as sns 
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
df_matches=pd.read_csv('matches.csv') 
df_matches
```

```python
df_leagues=pd.read_csv('leagues.csv') 
df_leagues
```

```python
df_matches['Country_name'] = df_matches['country_id'].map(df_leagues.set_index('country_id')['name.1'].drop_duplicates())
df_matches
```

```python
df_matches['League_name'] = df_matches['country_id'].map(df_leagues.set_index('country_id')['name'].drop_duplicates())
df_matches
```

```python
df_teams=pd.read_csv('teams.csv')
df_teams
```

```python
df_matches['Home_team_name'] = df_matches['home_team_api_id'].map(df_teams.set_index('team_api_id')['team_long_name'].drop_duplicates())
df_matches
```

```python
df_matches['Away_team_name'] = df_matches['away_team_api_id'].map(df_teams.set_index('team_api_id')['team_long_name'].drop_duplicates())
df_matches
```

```python
df_vizz=df_matches.groupby(['Country_name'])
df_vizz=df_vizz.first()
df_vizz
```

```python
plt.figure(figsize=(20,10))
sns.barplot(x='home_team_goal',y='Country_name',data=df_matches,palette='husl')
```

```python
plt.figure(figsize=(20,10))
sns.barplot(x='home_team_goal',y='Country_name',data=df_matches,palette='hls')
```

```python
plt.figure(figsize=(100,10)) 
sns.barplot(x='Home_team_name',y='away_team_goal',data=df_matches,palette='hls',ci=None) 
```

```python
df_vizz_teams=df_matches.groupby(['Home_team_name'])
df_vizz_teams=df_vizz_teams.first()
df_vizz_teams
```

```python
df_ta=pd.read_csv('team_attributes.csv')
df_ta 
```

```python
df_ta['team_name'] = df_ta['team_api_id'].map(df_teams.set_index('team_api_id')['team_long_name'].drop_duplicates())
df_ta
```

```python
df_ta['country'] = df_ta['team_api_id'].map(df_matches.set_index('home_team_api_id')['Country_name'].drop_duplicates())
```

```python
df_vizz_r=df_ta.groupby(['country'])
df_vizz_r=df_vizz_r.first() 
```

```python
plt.figure(figsize=(20,20))
sns.catplot(x='team_name',y='buildUpPlaySpeedClass',data=df_vizz_r,palette='hls') 
plt.xticks(rotation = 90) 
plt.figure(figsize=(20,20))
sns.catplot(x='team_name',y='buildUpPlayPassingClass',data=df_vizz_r,palette='hls')  
plt.xticks(rotation = 90)
```

```python
df_ta['home_goals'] = df_ta['team_api_id'].map(df_matches.set_index('home_team_api_id')['home_team_goal'].drop_duplicates()) 
sns.barplot(x='home_goals',y='buildUpPlaySpeedClass',palette='hls',data=df_ta)
```

```python
sns.barplot(x='home_goals',y='buildUpPlayPassingClass',palette='hls',data=df_ta)
```

```python
df_vizz_bu=df_ta.groupby(['buildUpPlayPassingClass','buildUpPlaySpeedClass']) 
df_vizz_bu=df_vizz_bu.first()
```

```python
fig, ax = plt.subplots(figsize=(15,7))
df_vizz_bu=df_ta.groupby(['buildUpPlayPassingClass','buildUpPlaySpeedClass']).count()['home_goals'].plot(ax=ax)
```
